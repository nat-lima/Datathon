```python
from pathlib import Path
import sqlite3
import pandas as pd
```

```python
caminho_db = Path.cwd() / "data" / "extraidos" / "dados.db"
print(caminho_db)
```

```python
conn = sqlite3.connect(caminho_db)
```

```python
consulta_tabelas = "SELECT name FROM sqlite_master WHERE type='table';"
tabelas = pd.read_sql_query(consulta_tabelas, conn)
print("📋 Tabelas disponíveis no banco:")
print(tabelas)
```

```python
df_applicants = pd.read_sql_query("SELECT * FROM applicants", conn)
df_prospects = pd.read_sql_query("SELECT * FROM prospects", conn)
df_vagas = pd.read_sql_query("SELECT * FROM vagas", conn)
```
